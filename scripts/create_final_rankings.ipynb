```python
# since there's not enough Polygon donors (just 160), we'll merge them with the Ethereum ecosystem donors

import pandas as pd

addressesETH = pd.read_csv('addressesETH.csv')
addressesPOLY = pd.read_csv('out/addresses_ranked_df.csv')
# drop first column in addressesPOLY (it's just a counter)
addressesPOLY = addressesPOLY.drop(columns=['Unnamed: 0'])
addressesPOLY
```

```python
# concatenate the two dataframes with no sorting
addresses = pd.concat([addressesPOLY, addressesETH], axis=0, ignore_index=True, sort=False)
# addresses = pd.concat([addressesETH, addressesPOLY], axis=0, sort=False)
addresses = addresses.head(1000)
addresses
```

```python
addresses.to_csv('FINAL_MERGED_RANKINGS.csv', index=False)
```
